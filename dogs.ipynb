```python
import pandas as pd
```

```python
df = pd.read_csv('dog_breeds_traits.csv')
print(df.head())
print(df.tail())
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(8,5))
plt.hist(df["size"].dropna(), bins=20, color="skyblue", edgecolor="black")
plt.grid(True)
plt.title("Distribuição da variável 'size'")
plt.xlabel("Size")
plt.ylabel("Frequência")
plt.show()
```
